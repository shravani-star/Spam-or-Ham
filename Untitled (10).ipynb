# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
import nltk
from wordcloud import WordCloud
nltk.download('stopwords')
from nltk.corpus import stopwords
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import joblib
```

```python
df = pd.read_csv("spam.csv", encoding='latin-1')[['v1', 'v2']]
df
```

```python
df.columns = ['label', 'text']
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
df.info()
```

```python
df.head()
```

```python
sns.countplot(x='label', data=df)
plt.xticks([0, 1], ['Not Spam', 'Spam'])
plt.title('Class Distribution')
plt.show()
```

```python
# percentage
spam_percentage = df['label'].value_counts(normalize=True) * 100
spam_percentage
```

```python
df['text_length'] = df['text'].apply(len)

plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='text_length', hue='label', bins=50, kde=True)
plt.title('Message Length Distribution by Class')
plt.xlabel('Length of Message')
plt.ylabel('Frequency')
plt.show()
```

```python
def clean_text(text):
    words = text.lower().translate(str.maketrans('', '', string.punctuation)).split()
    words = [word for word in words if word not in stopwords.words('english')]
    return " ".join(words)

df['clean_text'] = df['text'].apply(clean_text)

spam_words = " ".join(df[df['label'] == 1]['clean_text'])
ham_words = " ".join(df[df['label'] == 0]['clean_text'])

spam_wc = WordCloud(width=600, height=400, background_color='black').generate(spam_words)
ham_wc = WordCloud(width=600, height=400, background_color='white').generate(ham_words)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(spam_wc, interpolation='bilinear')
plt.title('Spam Word Cloud')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(ham_wc, interpolation='bilinear')
plt.title('Not Spam Word Cloud')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
df['tokens'] = df['clean_text'].apply(lambda x: x.split())

spam_tokens = [word for tokens in df[df['label'] == 1]['tokens'] for word in tokens]
ham_tokens = [word for tokens in df[df['label'] == 0]['tokens'] for word in tokens]

spam_freq = Counter(spam_tokens).most_common(20)
ham_freq = Counter(ham_tokens).most_common(20)

def plot_top_words(word_freq, title):
    words, freqs = zip(*word_freq)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(freqs), y=list(words))
    plt.title(title)
    plt.xlabel("Frequency")
    plt.ylabel("Word")
    plt.show()

plot_top_words(spam_freq, "Top 20 Words in Spam Emails")
plot_top_words(ham_freq, "Top 20 Words in Non-Spam Emails")
```

# Model Training

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
def predict_email(text):
    vec = vectorizer.transform([text])
    pred = model.predict(vec)[0]
    return "Spam" if pred == 1 else "Not Spam"

print(predict_email("Congratulations! You've won a free trip to Dubai!"))
```

```python
joblib.dump(model, 'spam_classifier.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```
